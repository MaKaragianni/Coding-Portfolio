```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
# Loading the data
insurance_data = pd.read_csv('insurance.csv',sep=',')
insurance_data.iloc[:,:]
```

```python
# Convert the 1D array to a 2D array
X = insurance_data['age'].values.reshape(-1, 1)
y = insurance_data['charges'].values
```

```python
# Scatter plot with age on the x-axis & charges on the y-axis
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color = 'b', alpha=0.5)
plt.title('Insurance Charges v/s Age')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.grid(True)
plt.show()
```

```python
# Creating & fitting linear regression model
model = LinearRegression()
model.fit(X, y)
```

```python
# Making predictions of the insurance charges (y) for a given age (x)
y_pred = model.predict(X)
y_pred
```

```python
# Scatter plot with the best-fit line
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='b', alpha=0.5, label='Actual data')
plt.plot(X, y_pred, color='r', linewidth=1, label='Best-fit line')
plt.title('Insurance charges v/s Age with Best-fit Line')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.legend()
plt.grid(True)
plt.show()
```
